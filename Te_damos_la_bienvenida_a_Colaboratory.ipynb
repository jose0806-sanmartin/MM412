¿Qué son las Bibliotecas en Python?


Las bibliotecas en Python son conjuntos de módulos, que son archivos que contienen funciones, clases y variables, diseñados para realizar tareas específicas. Estas bibliotecas proporcionan funcionalidades predefinidas para resolver problemas comunes de manera eficiente, evitando que tengas que escribir código desde cero.


Módulo: Es un archivo con código en Python, que puede contener funciones, variables y clases.


Biblioteca: Una biblioteca es un conjunto de módulos que agrupan funcionalidades relacionadas. Por ejemplo, una biblioteca puede contener varios módulos para operaciones matemáticas, gráficos o manejo de fechas.

math

Se utiliza para realizar operaciones matemáticas básicas con números escalares.

```
import math
x=16
#calcular la raiz cuadrada de x
resultado= math.sqrt(x)
print("la raiz cuadrada de 16 es:", resultado)
```

numpy

Se utiliza para trabajar con arreglos y realizar operaciones matemáticas de manera eficiente.

```
import numpy as np
arr = np.array([1, 2, 3, 4])
#realizar una operacion sobre array
squares = np.sqrt(arr) #raiz cuadrada de cada elemento
print("la raices cuadradas de los elementos son:", squares)
```

matplotlib.pyplot

Se utiliza para crear gráficos y visualizar datos.

```
import matplotlib.pyplot as plt
x= np.linspace(-10, 10, 100) # Generar 100 puntos de -10 a 10
y= x**2 #funcion cuadrática
plt.plot(x,y)
plt.title("gráfico de f(x)= X^2")
plt.xlabel("x")
plt.grid(True)
plt.show()
```

scipy

Permite realizar optimización y encontrar raíces de funciones.

```
import scipy.optimize as opt
def f(x):
  return x**2 -4 #raiz de la funcion es 2

#encontrar la raiz de la funcion usando el metodo de newton
root = opt.newton(f, 2) # estimacion inicial de 2
print("la raiz encontrada es:" , root)
```
